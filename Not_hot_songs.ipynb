```python
links = []
for i in range (1,16):
    start=str(i)
    if i ==1:
        url = "http://www.acclaimedmusic.net/061024/1948-02s.htm"
        links.append(url)
    else: 
        url = "http://www.acclaimedmusic.net/061024/1948-02s"+start+".htm"
        links.append(url)
    print(url)
links
```

```python
songs = []
artists = []

for url in links:
    response = requests.get(url)
    if ( response.status_code == 200 ):
        # parse all pages
        soup = BeautifulSoup(response.content, "html.parser")
        for i in range(0,len(soup.select("td a")),4):
            artists.append(soup.select("td a")[i].get_text())
        for i in range(1,len(soup.select("td a")),4):
            songs.append(soup.select("td a")[i].get_text())
        
print(len(songs))   
print(len(artists))   
```

```python
not_hot = pd.DataFrame({'Title':songs,'Artist':artists})
display(not_hot.shape)
display(not_hot.isna().sum())
not_hot.head()
```

```python
not_hot.to_csv('not_hot.csv',index=False)
```
